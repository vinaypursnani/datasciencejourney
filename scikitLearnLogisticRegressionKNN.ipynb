```python
# Logistic regression mainly used for categorical data
# we use a set of features to pridict the probablity of class membership
# algorith predicts the class of each observation, present in the dataset
# the equation is simplest logistic function used to perform logistic regression
```

```python
# the odds of the event is the ratio of the probablity of the event to its' compliment
# when you take logarithm of odds it becomes linear regression,
# so data generalisation switches from categorical to regression
```

```python
# ovr = one vs rest, most suitable for binary problems
# multinominal
# Specifies the norm used in penalization, newtonCG and LBFGS solves support only l2 penalty
# dual is only enabled for l2 penalty
# c is the inverse of regularization
# if warm_sart = True algo used previous solution of the call
```

```python
# choose an odd value of K for a 2 class problem i.e. binary classification to avoid the 
# case of "tied distance between two classes"
# K must not be a multiple of the number of classes
# the main drawback of KNN is the complexity in searching the nearest neighbours for each sample
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
type(iris_dataset)
```

```python
# describe dataset
print(iris_dataset['DESCR'])
```

```python
# feature_names
iris_dataset['feature_names'] # same as iris_dataset.feature_names
```

```python
# view target
iris_dataset['target']
```

```python
# view observations size
iris_dataset.data.shape # 150 observation 4 attributes
```

```python
# assign features data to X axis
X_features = iris_dataset.data
```

```python
# assign target data to y axis
Y_target = iris_dataset.target
```

```python
# view shapes of both axis
print(X_features.shape)
print(Y_target.shape)
```

```python
# import knn classifier method - import from sklearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
# instantiate the KNN estimator
knn = KNeighborsClassifier(n_neighbors=1) # so that the algo looks for neareast neighbour on start
```

```python
knn # should have printed whole KNN properties like in the next comment 
```

```python
# KNeighborsClassifier(algorithm='auto',
# leaf-size=20, metric='minkowski', metric_params=None, n_jobs=1, n_neighbors=1, p=2, weights='uniform')
```

```python
# fit the data that you have into KNN model (estimator)
knn.fit(X_features,Y_target)
```

```python
# lets create a new data to predict the outcome
X_new = [[3,5,4,1],[5,3,4,2]]
knn.predict(X_new)
```

```python
# now lets use logistic regression estimator
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()
```

```python
# fit data into the Logistic regression estimator
logReg.fit(X_features,Y_target)

# LogisticRegression (C=1.0, class_weight=None, dual=False, fit_intercept=True, intercept_scaling=1
# max_iter=100, multi_calss='ovr', n_jobs=1, penalty='12', random_state=None, solver='liblinear', tol=0.0001,
# )
```
