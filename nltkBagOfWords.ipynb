### Bag of Words covert text data into numerical feature vectors with a fized size
#### Step 1 - Tokenizing, assign a fixzed integer id to each word
#### Step 2 - Counting, number of ocurrences of each word
#### Step 3 - Store, store the value as feature in matrix format

```python
# goal is to convert a collection of text document, to a matrix of token count
# CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# instantiate the vectorizer
vectorizer = CountVectorizer()
```

```python
doc1 = 'Hello How are you'
doc2 = 'Welcome to the great great great India, bird of of gold'
doc3 = 'and have an Amazing Experience'
```

```python
listofdocuments = [doc1, doc2, doc3]
```

```python
#put the documents together
bag_of_words = vectorizer.fit(listofdocuments)
```

```python
bag_of_words
```

```python
# apply transform method
bag_of_words = vectorizer.transform(listofdocuments)
```

```python
print(bag_of_words) # tuple (document_number , feature_indices)  frequency of word
```

```python
# verify the vocabulary of repeated word
print(vectorizer.vocabulary_.get('great'))
print(vectorizer.vocabulary_.get('of'))
```
