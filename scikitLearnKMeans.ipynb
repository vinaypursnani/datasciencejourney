```python
# clustering is a group of similar data
# used to extract the structure of the data
# to identify groups in the data
# greater similarity between data points results in better clustering


# consider a scenario where each observed example has features,
# but has no lable or response attached to it
# in absence of a label we can identify, which datapoints in our dataset are similar
# each similar group of datapoints is called cluster

# Kmeans
# make some assumption through domain expertise then feed those assumptions to the algorithm
# Kmeans then randomly assigns a specific number of centroids to a data set
# it assigns datapoints to centroids, depending on their proximity to them
# once the datapoints are grouped seperately.
# Kmeans chosses a datapoint that is center of each group
# then it reassigns datapoints to these new centroids based on there proximity
# if this creats a new set group, then k means again chooses datpoint from center to make centroid
# this Iterate the process till the model is optimized
```

```python
# Kmeans in sklearn
```

```python
import numpy as np

# importing kmeans
from sklearn.cluster import KMeans

# dataset loading
from sklearn.datasets import make_blobs
```

```python
# define number of samples
n_samples = 300
# define random state value ot initialize the center
random_state = 20
# define number of features as 5
X,y = make_blobs(n_samples=n_samples,n_features=5,random_state=None)
# define number of clusters to be formed as 3 and in random state and fit features into the model
predict_y = KMeans(n_clusters=3,random_state=random_state).fit_predict(X)
# predict the estimator prediction
predict_y
# the results show which cluster the data belongs to
```
