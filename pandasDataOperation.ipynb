```python
import pandas as pd
```

```python
movie_rating = pd.DataFrame({'movie 1': [5,4,3,2,1,0], 'movie 2': [4,5,2,3,4,2]},
                            index = ['Varun','Vinay','Nash','Kevin','Gigi','Jim'])
movie_rating
```

```python
# df.applymap(function)

def movie_grade(rating):
    if rating==5:
        return 'A'
    if rating==4:
        return 'B'
    if rating==3:
        return 'C'
    if rating==2:
        return 'D'
    if rating ==1:
        return 'E'
    if rating ==0:
        return 'F'
    
movie_rating.applymap(movie_grade)
```

```python
# df.max(), df.min(), df.std()
df = pd.DataFrame({'Test1': [74,74,70,45,79], 'Test2':[29,27,24,25,30]},
                  index=['Reading','Listening','Speaking', 'Writing','Experimental'])
print(df)
print('df.max()\n', df.max())
print('df.min()\n', df.min())
print('df.std()\n', df.std())
```

```python
# standardizing dataset, enables us to analyze data set and catch outliers
def standardize_tests(test):
    return (test-test.mean()/test.std())
print(standardize_tests(df['Test1']))
print(standardize_tests(df['Test2']))

def standardize_test_scores(datafrm):
    return datafrm.apply(standardize_tests)
# apply can return both series or dataframe, applymap can only return a dataframe
standardize_test_scores(df) 
```

```python
import pandas as pd

df_student_math = pd.DataFrame({'student':['Tom','Jack','Dan','Ram','Jeff','David'],
                               'ID': [10,56,31,85,9,22]})
df_student_science = pd.DataFrame({'student': ['Tom','Ram','David'], 
                                  'ID': [10,12,22]})

# merge
pd.merge(df_student_math,df_student_science)
```

```python
# merge with key on student
pd.merge(df_student_math,df_student_science,on='student')
```

```python
# merge left join, on key ID and fill na values with X
pd.merge(df_student_math, df_student_science, on='ID', how='left').fillna('X')
```

```python
# concatenate dataframes
pd.concat([df_student_math, df_student_science], ignore_index=True)
```
